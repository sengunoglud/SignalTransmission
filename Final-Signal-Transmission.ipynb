```
data_cont = np.zeros((16,31))

for i in range(1,16):

    data_cont[i,0] = i
    # SIMON TASK
    text_1 = "data-1-"+str(i)+".csv"
    data = pd.read_csv(text_1)
    data_cont[i,1] = data["corr"].mean()
    data_cont[i,2] = data[data["type"]==1]["corr"].mean()
    data_cont[i,3] = data[data["type"]==2]["corr"].mean()
    data_cont[i,4] = data["rt"].mean()
    data_cont[i,5] = data[data["type"]==1]["rt"].mean()
    data_cont[i,6] = data[data["type"]==2]["rt"].mean()
    df = data
    z_scores = stats.zscore(df["rt"])

    abs_z_scores = np.abs(z_scores)
    df["abs_z_scores"] = abs_z_scores

    def remover(df, std=3):
       removed = (df < std)
       return removed

    removed = remover(df = df['abs_z_scores'], std=3)
    removeddata = df[removed]
    #removeddata

    # removed.shape
    # data.shape
    #1,2 done
    data1 = removeddata.groupby('type').describe()
    # print(data1)
    # data1
    # removeddata
    plt.style.use("fivethirtyeight")
    plt.plot(removeddata['trial'], removeddata['time'], c="b", linewidth=1)
    plt.xlabel("Trials")
    plt.ylabel("Time")
    plt.show()
    plt.style.use("fivethirtyeight")
    plt.plot(removeddata['trial'], removeddata['rt'],c="b", linewidth=1)
    plt.xlabel('Trials')
    plt.ylabel('RT')
    from scipy.signal import lfilter
    n = 10
    b = [1.0 / n] * n
    a = 1
    yy = lfilter(b, a, removeddata["rt"], axis=0)
    plt.plot(removeddata["trial"], yy, linestyle="-", c="b", linewidth=1)
    plt.xlabel("Trial")
    plt.ylabel("RT")
    print(data.dtypes)
    #contingency table
    pos_trial = pd.crosstab(removeddata['pos'], removeddata['trial'], margins=False)
    print(pos_trial)
    compatible_trial = pd.crosstab(removeddata['compatible'], removeddata['trial'], margins=False)
    print(compatible_trial)
    #sns.boxplot(removeddata["time"], color="blue", width=0.4, saturation=0.50, fliersize=3)
    outlier_vir = pd.DataFrame()
    from scipy.stats import norm
    outlier_vir['rt'] = data['rt']
    #sns.distplot(data['rt'], fit=norm)
    outlier_vir['rt'] = np.log(outlier_vir['rt'])
    fig, axs = plt.subplots(1,2)
    #sns.distplot(data['rt'], fit=norm, ax=axs[0])
    #sns.distplot(outlier_vir['rt'], fit=norm, ax=axs[1], axlabel="log")
    data_cont[i,7] = removeddata["corr"].mean()
    data_cont[i,8] = removeddata[removeddata["type"]==1]["corr"].mean()
    data_cont[i,9] = removeddata[removeddata["type"]==2]["corr"].mean()
    data_cont[i,10] = removeddata["rt"].mean()
    data_cont[i,11] = removeddata[removeddata["type"]==1]["rt"].mean()
    data_cont[i,12] = removeddata[removeddata["type"]==2]["rt"].mean()
    data = [['corrmean',removeddata["corr"].mean()],['type1corrmean',removeddata[removeddata["type"]==1]["corr"].mean()],['type2corrmean',removeddata[removeddata["type"]==2]["corr"].mean()], ['rtmean', removeddata["rt"].mean()], ['type1rtmean', removeddata[removeddata["type"]==1]["rt"].mean()], ['type2rtmean',removeddata[removeddata["type"]==2]["rt"].mean()]]
    df = pd.DataFrame(data,columns=['0','Values'])
    # df.to_csv('data.txt', sep='\t')
    #np.savetxt()

    ###################################################################################################################
    # GONOGO TASK
    # data = pd.read_csv('gonogo-besir.csv')
    text_2 = "data-2-" + str(i) +".csv"
    data2 = pd.read_csv(text_2)
    #data.head(15)
    # data.shape
    dropped = data2[data2['response'] != 'NONE']
    dropped.correctresponse.unique()
    dropped.type.unique()
    dropped.choice.unique()
    dropped.trial.shape
    dropped.info()
    described = dropped['rt'].describe()
    # described
    dropped1 = dropped
    abs_z_scores_2 = np.abs(stats.zscore(dropped1["rt"]))
    dropped1["abs_z_scores"] = abs_z_scores_2
    removed2 = remover(df = dropped1['abs_z_scores'], std=3)
    removeddata2 = dropped1[removed2]
    finaldata = removeddata2
    #data2.columns
    grbycorr = finaldata.groupby('correctresponse')['rt'].describe()
    #grbycorr
    pivot = finaldata.pivot_table('rt', index='responded', columns='correctresponse', aggfunc='mean')
    p_mean = pivot["P"]
    r_mean = pivot["R"]
    mean_mean = (p_mean + r_mean) / 2
    finaldata.stim.unique()
    stim_corr = finaldata[["stim", "corr"]]
    stim_corr.head(10)
    gr_stim_corr = stim_corr.groupby('stim')['corr'].value_counts()
    gr_stim_corr
    df_stim_corr = pd.DataFrame(gr_stim_corr)
    p_corr_ratio = df_stim_corr["corr"][0] / (df_stim_corr["corr"][1] + df_stim_corr["corr"][0]) * 100
    r_corr_ratio = df_stim_corr["corr"][2] / (df_stim_corr["corr"][2] + df_stim_corr["corr"][3]) * 100
    overall_corr_ratio = (p_corr_ratio + r_corr_ratio) / 2
    #data_cont[i,11] = gr_stim_corr
    # -*- coding: utf-8 -*-
    # 13. GoNogo - OutElim Incog/Inhibition Correctness Ratio
    # 14. GoNogo - OutElim Congruent Correctness Ratio
    # 15. GoNogo - OutElim Congruent Correct Mean Resp Time
    # 16. GoNogo - OutElim Overall Correctness Ratio
    
    data_cont[i,13] = p_mean
    data_cont[i,14] = r_mean
    data_cont[i,15] = mean_mean
    data_cont[i,16] = p_corr_ratio
    data_cont[i,17] = r_corr_ratio
    data_cont[i,18]= overall_corr_ratio

    ###################################################################################################################
    # SignalTransmission
    # data = pd.read_csv("SignalTransmission.csv")
    # data_3 = pd.read_csv("SignalTransmission.csv")
    text_3 = "data-3-" + str(i)+".xlsx"
    data3 = pd.read_excel(text_3)
    data_cont[i,19] = data3[data3["IsCorrect"]==True]["responseTime"].mean()
    data4 = data3.loc[:,['level','_id']].groupby('_id').describe()
    state_iscorrect = data3.loc[:, ['state', 'IsCorrect']]
    # state_iscorrect
    by_state = state_iscorrect.groupby('state').count()
    by_iscorrect = state_iscorrect.groupby('IsCorrect').count()
    # by_state
    # by_iscorrect
    #data[data["IsCorrect"]==True][data["state"] == 'Normal'].count()
    x = data3.groupby('state').IsCorrect.agg(['sum'])
    x['state'] = by_state
    x['ratio'] = (x['sum'] / x['state']) * 100
    normal = x['ratio'][0]
    reverse = x['ratio'][1]
    normal_corr_ratio = normal
    reverse_corr_ratio = reverse
    overall_corr_ratio1 = (normal + reverse) / 2 
    # data3.columns
    video_audio_rt = data3.loc[:, ['targetStimuli', 'IsCorrect']]
    # video_audio_rt
    video_audio_rt.targetStimuli.unique()
    dropped_vart = video_audio_rt[video_audio_rt['targetStimuli'] != 'Parasite']
    # dropped_vart.shape
    by_target = dropped_vart.groupby('targetStimuli').count()
    by_iscorrect1 = dropped_vart.groupby('IsCorrect').count()
    y = dropped_vart.groupby('targetStimuli').IsCorrect.agg(['sum'])
    #by_target
    #by_iscorrect1
    y['targer'] = by_target
    y['ratio'] = (y['sum'] / y['targer']) * 100
    target_rt_ = data3.loc[:, ['targetStimuli', 'responseTime','IsCorrect']]
    dropped_target_rt_ = target_rt_[target_rt_['targetStimuli'] != 'Parasite']
    dropped1_target_rt_ = dropped_target_rt_[dropped_target_rt_['IsCorrect'] != False]
    dropped1_target_rt_.IsCorrect.unique()
    data_cont[i,20] = target_rt_audio = dropped1_target_rt_[dropped1_target_rt_["targetStimuli"]=='Audio']["responseTime"].mean()
    data_cont[i,21] = target_rt_video = dropped1_target_rt_[dropped1_target_rt_["targetStimuli"]=='Video']["responseTime"].mean()
    dropped_data = data3[data3['targetStimuli'] != 'Parasite']
    pb_ic_rt = dropped_data.loc[:, ['pressedButton', 'IsCorrect','responseTime']]
    by_button = pb_ic_rt.groupby('pressedButton').count()
    by_iscorrect2 = pb_ic_rt.groupby('IsCorrect').count()
    z = pb_ic_rt.groupby('pressedButton').IsCorrect.agg(['sum'])
    dropped_by_button = by_button.drop('responseTime', axis=1)
    z['isCorrect'] = dropped_by_button
    z['ratio'] = (z['sum'] / z['isCorrect']) * 100
    trues_only = pb_ic_rt[pb_ic_rt['IsCorrect'] != False]
    data_cont[i,22] = right = trues_only[trues_only["pressedButton"] == 'Right']["responseTime"].mean()
    data_cont[i,23] = left = trues_only[trues_only["pressedButton"] == 'Left']["responseTime"].mean()
    state_iscorrect
    
    #data.columns
    # 17. SignalTransmission - Correct Mean Response Time
    # 18. SignalTransmission - Audio ResponseTime ortalaması
    # 19. SignalTransmission - Video ResponseTime ortalaması 
    # 20. SignalTransmission - Right-Button ResponseTime ortalaması
    # 21. SignalTransmission - Left-Button ResponseTime ortalaması
    # 22. SignalTransmission - EndGame Level
    # 23. SignalTransmission - Overall Correctness Ratio
    # 24. SignalTransmission - Normal State Correctness Ratio
    # 25. SignalTransmission - Reverse State Correctness Ratio 
    
    data_cont[i,24] = right = trues_only[trues_only["pressedButton"] == 'Right']["responseTime"].mean()
    data_cont[i,25] = left = trues_only[trues_only["pressedButton"] == 'Left']["responseTime"].mean()
    data_cont[i,26] = normal_corr_ratio
    data_cont[i,27] = reverse_corr_ratio
    data_cont[i,28] = overall_corr_ratio1
    # data_cont[i,29] = changed_ratio
    # data_cont[i,30] = stable_ratio

    data_d = data3[data3['targetStimuli'] != 'Parasite']
    t = data_d.loc[:, ['state', 'IsCorrect']]
    temp = ""
    t["Status"] = "None"
    for i in range(len(t.index)):
      if i == len(t) - 1:
        t["Status"].iloc[i] = "Stable"
        break
      if t["state"].iloc[i] != t["state"].iloc[i+1]:
        temp = "Changed"
      else:
        temp = "Stable"
      t["Status"].iloc[i] = temp

    print(t)
    by_status = t.groupby('Status').count()
    by_iscorrect3 = t.groupby('IsCorrect').count()
    q = t.groupby('Status').IsCorrect.agg(['sum'])
    dropped_by_status = by_status.drop(['IsCorrect'], axis=1)
    q["Total"] = dropped_by_status
    q["ratio"] = (q["sum"] / q["Total"]) * 100
    changed_ratio = q["ratio"][0]
    stable_ratio = q["ratio"][1]

    """data_cont[i,21] = x
    data_cont[i, 22] = y
    data_cont[i,23] = target_rt_audio
    data_cont[i,24] = target_rt_video
    data_cont[i,25] = z
    data_cont[i,26] = right
    data_cont[i,27] = left
    data_cont[i,28] = q"""
```
